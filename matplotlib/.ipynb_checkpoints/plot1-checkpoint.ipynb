```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x=np.linspace(0,5,11)
y=x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,'g-')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Demo Data')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
fig=plt.figure()

axes=fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('x axis')
axes.set_ylabel('y axis')
axes.set_title('Demo Data')
```

```python
fig=plt.figure()

axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes2.plot(y,x,'r')
axes2.set_title('smaller plot')
axes1.set_title('larger plot')
```

```python
fig,axes= plt.subplots(nrows=3,ncols=3)
plt.tight_layout()
```

```python
fig,axes= plt.subplots(nrows=1,ncols=2)

# for current_ax in axes:
#     current_ax.plot(x,y)

axes[0].plot(x,y)
axes[0].set_title('first plot')
axes[1].plot(y,x)
axes[1].set_title('second plot')
```

```python
axes
```

```python
# fig=plt.figure(figsize=(3,2),dpi=100)
fig=plt.figure(figsize=(8,2))
ax= fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(x,y)
```

```python
fig.savefig('graph_pic.png',dpi=200)
```

```python
fig=plt.figure()
ax= fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label="X Sqaured")
ax.plot(x,x**3,label="X Cubed")

# ax.legend(loc=10)
ax.legend(loc=(0.1,0.1))
```

```python
fig=plt.figure()
ax= fig.add_axes([0,0,1,1])
# ax.plot(x,y,color='#000',linewidth=100,alpha=0.5)
# ax.plot(x,y,color='#000',lw=100,alpha=0.5)
ax.plot(x,y,color='#000',linewidth=3,linestyle="-",marker='o',markersize=20,markerfacecolor='green',markeredgewidth=3,markeredgecolor='blue')
```

```python
fig=plt.figure()
ax= fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',linewidth=3,linestyle="--")
ax.set_xlim([0,1])
ax.set_ylim([0,1])
```
