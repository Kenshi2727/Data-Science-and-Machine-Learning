```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)# by deafault estimator is set to mean
```

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)#now extimator is set to standard deviation
```

```python
sns.countplot(x='sex',data=tips) #estimator is counting no. of occurences
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=False)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex',dodge=True) #split deprecated use dodge instead
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips)#comination of strip and violin plots
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='bar') #factorplot deprecated
```
