```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.set_style('whitegrid')
sns.countplot(x='sex',data=tips)
```

```python
tips=sns.load_dataset('tips')
sns.countplot(x='sex',data=tips)
sns.despine(left=True,bottom=True) #by default right and upper spines are true
```

```python
plt.figure(figsize=(12,3))
sns.countplot(x='sex',data=tips)
```

```python
sns.set_context('poster',font_scale=3)#3 times the deafult
sns.countplot(x='sex',data=tips)
```

```python
sns.set_context('notebook')
sns.countplot(x='sex',data=tips)
```
