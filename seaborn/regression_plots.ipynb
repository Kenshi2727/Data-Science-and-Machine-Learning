```python
import seaborn as sns
%matplotlib inline
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'],scatter_kws={'s':100})
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',row='time',hue='sex')#separated by columns and rows
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',
          aspect=0.6,height=8)  #size is deprecated use height instead
```
